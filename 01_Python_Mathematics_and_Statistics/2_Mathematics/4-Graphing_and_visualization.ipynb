# Graphing and visualization(Grafikleme ve görselleştirme)

### matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
x = 4
y = 6
plt.plot(x,y) # plot bize grafigi cizdi
plt.plot(x,y,'ro') # kesisme noktasini verir
plt.axis('square') # grafigi kare sekline getirdik
plt.axis([-6,6,-8,8]) # grafigimizi istedimiz sayi araliklarinda gosterilmesini sagladik
plt.grid()
plt.show()

```

```python
x = [-1,6,-8,10,5,-5,9]
y = [-1,-6,-9,7,8,9,1]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o', label= f'{i+1}') # etiket attik

plt.axis('square')
plt.axis([-12,12,-12,12])
plt.legend()# hangi rengin ne oldufunu grafikta gosterir
plt.grid()
plt.show()    
```

```python
# koordinat duzlemini kendimize gore ayarlama
plt .plot(5,1,'r v') # deger ve kestikleri noktanin simgesini gosterdik
axis = plt.gca()
ylim = axis.get_ylim() # get_ylim ile mevcut koordinat boyutlarini getirdik
print(ylim)
axis.set_ylim([ylim[0],6]) # set_ylim ile yeni deger atamasi
plt.show()
```

```python
# bir fonksiyon sorusu cozelim
import sympy as sym # matematiksel gosterim icin sympy modulumuzu ekledik

x = sym.symbols('x')
y = 2*x**2 - 3*x

for i in range(-12,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x koordinat')
plt.ylabel(f'f(x) = {y}')
plt.show()


```

```python
# professional graphing lines -- profesyonel bir grafik cizelim
x = 2
y = 4 
plt.plot(x,y,'r o')
plt.plot([0,x],[0,y], 'r') # orjinden gececek sekilde cizgi cektik

plt.axis('square') # duzlemimizi kesisim yerinde kare olmasini sagladik
plt.axis([-6,6,-6,6]) # koordinatimizi x ve y seklinde sinir araligini koyduk
plt.grid() # matemaktiksel defter stili grid yaptik

plt.plot([-6,6],[0,0],'k') # x eksenini siyah cizgi cek
plt.plot([-0,0],[-6,6],'k') # y eksenine siyah cizgi cek dedik
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = range(-20, 21)

plt.figure(figsize=(12, 6))# "12 birim genişlik, 6 birim yükseklikte YENİ bir grafik penceresi oluştur"

# Orijinal kodun çizdiği çizgiler
for i in x:
    plt.plot([0, i], [0, abs(i)**(1/2)], 'b-', alpha=0.3, linewidth=0.5)

# Fonksiyonun kendisini de çizelim
x_vals = np.linspace(-20, 20, 1000)
y_vals = np.sqrt(np.abs(x_vals))
plt.plot(x_vals, y_vals, 'r-', linewidth=2, label='f(x) = √|x|')

plt.xlabel('x')
plt.ylabel('f(x) = √|x|')
plt.title('Karekök Fonksiyonu ve Orijinden Çizgiler')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# DAHA AKILLI KUTU ÇİZİMİ:
def kutu_ciz(x1, y1, x2, y2):
    """(x1,y1) sol alt, (x2,y2) sağ üst köşe olan kutu çizer"""
    
    # Köşe noktaları
    x_koseler = [x1, x2, x2, x1, x1]  # Kapalı kutu için ilk noktaya dön
    y_koseler = [y1, y1, y2, y2, y1]
    
    plt.plot(x_koseler, y_koseler, 'r-', linewidth=3)
    plt.fill_between([x1, x2], y1, y2, alpha=0.2, color='blue')  # İçini boya

# Kutu çiz
kutu_ciz(0, 0, 2, 2)

plt.title('Dolu Kutu')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()
```
