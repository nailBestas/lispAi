```python
import matplotlib.pyplot as plt

# DAHA AKILLI KUTU ÇİZİMİ:
def kutu_ciz(x1, y1, x2, y2):
    """(x1,y1) sol alt, (x2,y2) sağ üst köşe olan kutu çizer"""
    
    # Köşe noktaları
    x_koseler = [x1, x2, x2, x1, x1]  # Kapalı kutu için ilk noktaya dön
    y_koseler = [y1, y1, y2, y2, y1]
    
    plt.plot(x_koseler, y_koseler, 'r-', linewidth=3)
    plt.fill_between([x1, x2], y1, y2, alpha=0.2, color='blue')  # İçini boya

# Kutu çiz
kutu_ciz(0, 0, 2, 2)

plt.title('Dolu Kutu')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()
```

```python
# coklu karmasik problem

# 1-Kutuphanelri ekle

import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

# 2- Sembolleri tanimlayalim

x, a = sym.symbols('x a')

# 3-Fonksiyonumuzu yazalim
# Orijinal fonksiyon: y = a / (x² - a)
y = a / (x**2 - a)
print("Fonksiyon:", y)

# 4- Faarkli degerler icin fonksiyonumuzu hesaplayalim
# a = 1, 2, 3, 4 için fonksiyonlar
a_values = [1, 2, 3, 4]
functions = []

for a_val in a_values:
    y_sub = y.subs(a, a_val)  # a yerine değerini koy
    functions.append(y_sub)
    print(f"a = {a_val}: y = {y_sub}")


 # 5- Grafikleri cizelim

print("\n--- SymPy ile Grafikler ---")
p = None
for i, a_val in enumerate(a_values):
    y_func = y.subs(a, a_val)
    if i == 0:
        p = sym.plot(y_func, (x, -5, 5), show=False, 
                    label=f'a={a_val}', line_color=['red','blue','green','orange'][i])
    else:
        p_temp = sym.plot(y_func, (x, -5, 5), show=False, 
                         label=f'a={a_val}', line_color=['red','blue','green','orange'][i])
        p.append(p_temp[0])

p.title = 'y = a / (x² - a)'
p.ylim = (-10, 10)
p.legend = True
p.show()   


# 6-Matplotlib ile cizim
print("\n--- Matplotlib ile Detaylı Grafik ---")
plt.figure(figsize=(12, 8))
colors = ['red', 'blue', 'green', 'orange']

for i, a_val in enumerate(a_values):
    # x değerlerini oluştur (sonsuzluk noktalarını atla)
    x_vals = np.linspace(-5, 5, 1000)
    
    # ±√a noktalarını çıkar (payda sıfır olan yerler)
    sqrt_a = np.sqrt(a_val)
    mask = (np.abs(x_vals - sqrt_a) > 0.05) & (np.abs(x_vals + sqrt_a) > 0.05)
    x_vals = x_vals[mask]
    
    # y değerlerini hesapla
    y_vals = a_val / (x_vals**2 - a_val)
    
    # Grafiği çiz
    plt.plot(x_vals, y_vals, color=colors[i], linewidth=2, label=f'a = {a_val}')
    
    # Dikey asimptotları çiz
    plt.axvline(x=sqrt_a, color=colors[i], linestyle='--', alpha=0.5)
    plt.axvline(x=-sqrt_a, color=colors[i], linestyle='--', alpha=0.5)

plt.axhline(y=0, color='black', alpha=0.3)
plt.axvline(x=0, color='black', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = a / (x² - a)  (Farklı a değerleri için)')
plt.legend()
plt.ylim(-20, 20)
plt.xlim(-5, 5)
plt.show()

# 7-Kritik noktalari hesaplayalim 
# In [7]:
print("\n--- Kritik Noktalar ---")
for a_val in a_values:
    sqrt_a = np.sqrt(a_val)
    print(f"a = {a_val}:")
    print(f"  • Payda sıfır olduğu noktalar: x = ±√{a_val} ≈ ±{sqrt_a:.2f}")
    print(f"  • Dikey asimptotlar: x = {sqrt_a:.2f} ve x = -{sqrt_a:.2f}")
    print(f"  • x → ∞ için y → 0 (Yatay asimptot: y = 0)")
    print()




# 8-Ozel degerleri hesaplama
# In [8]:
print("\n--- Özel Noktalar ---")
test_points = [-4, -2, 0, 2, 4]

for a_val in a_values:
    print(f"\na = {a_val}:")
    y_func = y.subs(a, a_val)
    
    for point in test_points:
        if abs(point) != np.sqrt(a_val):  # Sonsuzluk noktalarını atla
            y_val = y_func.subs(x, point)
            print(f"  x = {point:2} → y = {float(y_val):7.2f}")
```

```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt

# create a matrix-matrisimizi olusturuyoruz
#2x2'lik bir matris oluşturuyoruz
# A = [ 2  2 ]
 #   [ 1  4 ]
A = [ [2,2],[1,4] ]

# show it mathematically
display(Math(sym.latex(sym.sympify(np.array(A)))))
'''
np.array(A) → listeyi numpy array'e çevir

sym.sympify() → numpy array'ini sympy formatına çevir

sym.latex() → LaTeX formatına dönüştür

Math() → LaTeX'i matematiksel gösterime hazırla

display() → Jupyter'de güzel göster
'''

# now image it
plt.imshow(A) #Matrisi renkli bir görsel olarak çiziyoruz. Her sayı farklı renkte gözükür.

# CORRECTED axis ticks
# Eksen etiketlerini ayarlıyoruz
plt.xticks([0,1])
plt.yticks([0,1])  # 

# Add colorbar and labels for better visualization
plt.colorbar()
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.title('Matrix Visualization')

plt.show()
```

```python
# Matrix ve goruntu elde etme
# Making images from matrices

A = np.zeros((10,14))

print( np.shape(A) )


#Bu nested loop (iç içe döngü)
for i in range(0,np.shape(A)[0]):
  for j in range(0,np.shape(A)[1]):

    # populate the matrix
    A[i,j] = 3*i-4*j

print(A)
plt.imshow(A)
plt.plot([0,3],[8,2],'r',linewidth=4)
plt.set_cmap('Purples')

for i in range(0,np.shape(A)[0]):
  for j in range(0,np.shape(A)[1]):
    plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')


plt.show()
```
