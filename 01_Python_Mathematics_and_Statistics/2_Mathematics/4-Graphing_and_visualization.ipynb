### matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
x = 4
y = 6
plt.plot(x,y) # plot bize grafigi cizdi
plt.plot(x,y,'ro') # kesisme noktasini verir
plt.axis('square') # grafigi kare sekline getirdik
plt.axis([-6,6,-8,8]) # grafigimizi istedimiz sayi araliklarinda gosterilmesini sagladik
plt.grid()
plt.show()

```

```python
x = [-1,6,-8,10,5,-5,9]
y = [-1,-6,-9,7,8,9,1]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o', label= f'{i+1}') # etiket attik

plt.axis('square')
plt.axis([-12,12,-12,12])
plt.legend()# hangi rengin ne oldufunu grafikta gosterir
plt.grid()
plt.show()    
```

```python
# bir fonksiyon sorusu cozelim
import sympy as sym # matematiksel gosterim icin sympy modulumuzu ekledik

x = sym.symbols('x')
y = 2*x**2 - 3*x

for i in range(-12,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x koordinat')
plt.ylabel(f'f(x) = {y}')
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

x = range(-20, 21)

plt.figure(figsize=(12, 6))# "12 birim genişlik, 6 birim yükseklikte YENİ bir grafik penceresi oluştur"

# Orijinal kodun çizdiği çizgiler
for i in x:
    plt.plot([0, i], [0, abs(i)**(1/2)], 'b-', alpha=0.3, linewidth=0.5)

# Fonksiyonun kendisini de çizelim
x_vals = np.linspace(-20, 20, 1000)
y_vals = np.sqrt(np.abs(x_vals))
plt.plot(x_vals, y_vals, 'r-', linewidth=2, label='f(x) = √|x|')

plt.xlabel('x')
plt.ylabel('f(x) = √|x|')
plt.title('Karekök Fonksiyonu ve Orijinden Çizgiler')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# DAHA AKILLI KUTU ÇİZİMİ:
def kutu_ciz(x1, y1, x2, y2):
    """(x1,y1) sol alt, (x2,y2) sağ üst köşe olan kutu çizer"""
    
    # Köşe noktaları
    x_koseler = [x1, x2, x2, x1, x1]  # Kapalı kutu için ilk noktaya dön
    y_koseler = [y1, y1, y2, y2, y1]
    
    plt.plot(x_koseler, y_koseler, 'r-', linewidth=3)
    plt.fill_between([x1, x2], y1, y2, alpha=0.2, color='blue')  # İçini boya

# Kutu çiz
kutu_ciz(0, 0, 2, 2)

plt.title('Dolu Kutu')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()
```

```python
# Lines in slope-intercept form
# Eğim-kesişim formundaki çizgiler  y =mx+b  m =egim/slope demek  b=kesim /intercept 

import matplotlib.pyplot as plt

# y = mx+b egim

x = [-4,4]
m = 2
b = 3

y = [0,0]
for i in range(len(x)):
    y[i] = m*x[i] + b

print(x,y)

plt.plot(x,y,label = (f'y = {m}x +{b}'))
plt.legend()

plt.plot(x,y)
plt.axis('square')
plt.grid()
plt.xlim(x) # x sinirina kadar doldurulmasini istiyopruz
plt.ylim(x) # sinirina kadar doldurulmasini istiyoruz
axis = plt.gca() # "Get Current Axis" - Mevcut eksenleri al
plt.plot(axis.get_xlim(),[0,0], 'k--')
plt.plot([0,0],axis.get_ylim(), 'k--')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veriler
x = [-3, -2, -1, 0, 1, 2, 3]
m, b = 2, 1
y = [m*xi + b for xi in x]  # y = 2x + 1

# Grafik
plt.plot(x, y, 'r-', linewidth=2, label=f'y = mx + b ({m}, {b})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Doğru Grafiği')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # x ekseni
plt.axvline(0, color='black', linewidth=0.5)  # y ekseni
plt.legend()  # LABEL İÇİN ŞART!
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x değerlerini oluştur (-4'ten 4'e kadar)
x = np.linspace(-4, 4, 100)  # -4 ile 4 arasında 100 nokta

# İki doğrunun denklemleri
y1 = 0.5*x + 2      # y = 0.7x + 2
y2 = -3*x + 0.25 # y = -1.25x + 0.75

# Grafiği oluştur
plt.figure(figsize=(10, 6))

# Doğruları çiz
plt.plot(x, y1, 'b-', linewidth=2, label='y = 0.7x + 2')
plt.plot(x, y2, 'r-', linewidth=2, label='y = -1.25x + 0.75')

# Eksenleri çiz
plt.axhline(0, color='black', linewidth=0.5)  # x ekseni
plt.axvline(0, color='black', linewidth=0.5)  # y ekseni

# Eksen sınırlarını ayarla
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# Grid ve etiketler
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('İki Doğrunun Grafiği')
plt.legend()

# Kesim noktasını bulalım (isteğe bağlı)
# 0.7x + 2 = -1.25x + 0.75
kesim_x = (0.75 - 2) / (0.7 - (-1.25))
kesim_y = 0.7*kesim_x + 2
plt.plot(kesim_x, kesim_y, 'go', markersize=8, label=f'Kesim noktası: ({kesim_x:.2f}, {kesim_y:.2f})')

plt.legend()
plt.show()

# Kesim noktasını yazdıralım
print(f"Kesim noktası: x = {kesim_x:.2f}, y = {kesim_y:.2f}")
```

```python
# Graphing rational functions --rasyonel fonksiyonlar yani polinomlarin orani diyebiliriz

import numpy as np
import matplotlib.pyplot as plt

# y = 3-x**2 polinomunu veya fonsiyonunu cozelim

# x degerlerini yapalim
x = np.linspace(-4,4,10) # np.linspace(başlangıç, bitiş, nokta_sayısı)

# y degerini girelim
y = np.zeros(len(x))# x elmanina bagli oalrak o kadar sifir atar yani 10 tane 0 atacaktir

# deger atamasi icin dongu olusturalim
for i in range(len(x)):
    y[i] = 3-x[i]**2

plt.plot(x,y,'s-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x değerlerini oluştur
x = np.linspace(-2, 2, 400)

# Polinom fonksiyonları
y1 = x**2
y2 = x**3
y3 = x**4

# Egrileri çiz
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='$y = x^2$', color='blue', linewidth=2)
plt.plot(x, y2, label='$y = x^3$', color='red', linewidth=2)
plt.plot(x, y3, label='$y = x^4$', color='green', linewidth=2)

# Grafik ayarları
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polinom Fonksiyonlarının Grafikleri')
plt.legend()
plt.grid(True)
plt.ylim(-1, 16)
plt.xlim(-2.5, 2.5)

# Grafiği göster
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x değerlerini oluştur
x = np.linspace(-2, 2, 400)

# Doğrusal (lineer) fonksiyonlar
y1 = x      # y = x
y2 = 2 * x  # y = 2x
y3 = 3 * x  # y = 3x

# Grafiği çiz
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='$y = x$', color='blue', linewidth=2)
plt.plot(x, y2, label='$y = 2x$', color='red', linewidth=2)
plt.plot(x, y3, label='$y = 3x$', color='green', linewidth=2)

# Grafik ayarları
plt.xlabel('x')
plt.ylabel('y')
plt.title('Doğrusal Fonksiyonların Grafikleri')
plt.legend()
plt.grid(True)
plt.ylim(-6, 6)  # Doğrular için uygun y aralığı
plt.xlim(-2.5, 2.5)

# Grafiği göster
plt.show()
```

```python
# Plotting with Sympy
# sympy modulu ile cizim
import matplotlib.pyplot as plt
# modulumuzu ice aktaralim
import sympy as sym 
# sembolleri olusturmanin diger yolu modul olarak ekleme x sembolunu olsuturalim
#x = sym.symbols('x') diger yol
from sympy.abc import x 

y = 2*x**2
p = sym.plot(y,show=False) # grafigi kapattik p artık çizgi listesi
p.xlim = [0,20]
p[0].line_color = 'r' #  # p listesindeki İLK çizgi (0. index)
plt.title(f'Bu bir $y = {sym.latex(2*x**2)}$ fonksiyonudur')  # $ $ = "Bu matematik formülüdür, onu güzel göster"

#SymPy'in latex() fonksiyonu bize LaTeX kodu veriyor, biz de onu $ $ içine koyuyoruz ki matplotlib anlayabilsin!

p.show() # tekrar actik



```

```python
import matplotlib.pyplot as plt

# Birden fazla çizgi çiz
x = [1, 2, 3, 4]
y1 = [1, 4, 2, 3]
y2 = [2, 3, 1, 4]

# Çizgileri çiz ve listeye kaydet
lines = plt.plot(x, y1, x, y2)

# lines[0] = ilk çizgi (y1), lines[1] = ikinci çizgi (y2)

# Çizgilerin renklerini değiştir
lines[0].set_color('r')  # İlk çizgiyi KIRMIZI yap
lines[1].set_color('g')  # İkinci çizgiyi YEŞİL yap

# Diğer özellikleri de değiştirebilirsin
lines[0].set_linewidth(3)  # Kalınlık
lines[1].set_linestyle('--')  # Kesikli çizgi

plt.show()
```
